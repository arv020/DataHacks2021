```python
def preprocessor(df1):
    df1 = df1.replace({"***":np.NaN, '--':np.NaN})
    df1 = df1.dropna(how = "all", axis = 1)
    df1 = df1.drop([df1.columns[0], df1.columns[1], df1.columns[2], df1.columns[4]], axis =1)
    df1 = df1.rename({df1.columns[0]:'Year'})
    df1 = df1.drop([i for i in df1.columns if 'year' in i], axis =1)
    df1 = df1.dropna()
    return df1
    
```

```python
def preprocessor_test(df):
    df = df.replace({"***":np.NaN, '--':np.NaN})
    df = df.dropna(how = "all", axis = 1)
    df = df.drop([df.columns[0], df.columns[1], df.columns[2]], axis =1)
    df = df.rename({df.columns[0]:'Year'})
    df = df.drop([i for i in df.columns if 'year' in i], axis =1)
    return df
```
