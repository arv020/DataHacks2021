```python
def extractor(df):
    df = df.replace("***", np.NaN)
    df = df.dropna(how = 'all', axis =1 )
    df = df.rename({'econ_year': 'Year'}, axis = 1)
    df = df.drop([i for i in econ2.columns if 'year' in i], axis=1)
    growths = []
    for i in df.iloc[:,0].unique():
        
        growth = df.loc[df.iloc[:,0] == i].iloc[:,5].dropna()
        
        growth_rate = (growth.iloc[-1] - growth.iloc[0])/growth.iloc[0]
        
        growths.append(growth_rate)
    return pd.Series(growths)
```
